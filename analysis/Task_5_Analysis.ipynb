```python
#bringing in my cleaned data 
import pandas as pd
from scripts import project_functions
data = pd.read_excel('scripts/data.xlsx')
expr_df = project_functions.load_and_process(data)
```

```python
# filter out non-expressed genes
expr_df = expr_df.loc[expr_df.sum(axis=1) > 0, :]

# filter out lowly expressed 
expr_df = expr_df.loc[expr_df.sum(axis=1) > 25, :]
```

```python
# log transform
import numpy as np
expr_df['Log Stationary Phase'] = np.log(expr_df['Avg. Stationary Phase']) 
expr_df['Log Growth in Media (3h)'] = np.log(expr_df['Avg. Growth in Media (3h)']) 
expr_df['Log Media and Ampicillin (3h)'] = np.log(expr_df['Avg. Media and Ampicillin (3h)']) 

# drop coulmns of averages 
new_df = expr_df.drop(columns=['Avg. Stationary Phase', 'Avg. Growth in Media (3h)', 'Avg. Media and Ampicillin (3h)'])
```

```python
new_df
```

```python
# filtering out negative values caused by log transformation as these genes can also be considered lowly expressed ie log 0.5 = -1 
plotdata = new_df[new_df.select_dtypes(include=[np.number]).ge(0).all(1)]
plotdata 
```

```python
# now lets make a histogram 
hist = plotdata.hist(bins=5,figsize =(10,10))
hist
# log stationary phase is still skewed right but this looks much better 
```

```python
# i want to compare genes from the fresh media to the media plus ampicillin treatment 
# goal is to see which genes are up regulated during treatment with ampicilin 
# first i will subtract fresh media from ampicillin then visualize

plotdata['Ampicillin Remove Media'] = plotdata["Log Media and Ampicillin (3h)"] - plotdata["Log Growth in Media (3h)"]
plotdata 
```

```python
# visualize the data
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))

ax = sns.countplot(data=plotdata, x = "Gene Symbol", order=plotdata["Gene Symbol"].value_counts().index[:15])

plt.title("Top 25 Upregulated Genes in E. coli during Treatment with Ampicillin")
plt.ylabel("Count")
plt.show()
```

```python
# lets look at a couple more
plt.figure(figsize=(15,15))
ax = sns.countplot(data=plotdata, x = "Gene Symbol", order=plotdata["Gene Symbol"].value_counts().index[:25])

plt.title("Top 25 Upregulated Genes in E. coli during Treatment with Ampicillin")
plt.ylabel("Count")
plt.show()
# top upregulated genes are copA and dnaX
```

```python
# make a subset with the top genes of interest so this dataset is easier to work with 
dff = plotdata.sort_values(['Ampicillin Remove Media'], ascending=False)
dff = dff[:30]
dff
# scatter plot
plt.figure(figsize=(25,25))
ax = sns.scatterplot(data=dff, x="Gene Symbol", y="Ampicillin Remove Media")
plt.xticks(plt.xticks()[0], rotation=90)
plt.title("Upregulated Genes in E. coli during Treatment with Ampicillin")
plt.show()
```

```python
# make a subset with the top genes of interest so this dataset is easier to work with 
dff1 = plotdata.sort_values(['Ampicillin Remove Media'], ascending=False)
dff1 = dff1[:10]
dff1
#boxplot
plt.figure(figsize=(10,10))
sns.boxplot(x='Gene Symbol',y='Ampicillin Remove Media',data=dff1)
plt.xticks(plt.xticks()[0], rotation=90)
plt.title("Upregulated Genes in E. coli during Treatment with Ampicillin")
plt.show()
```
